```python
import pandas as pd
import csv
import numpy as np
```

```python
data_raw = pd.read_csv('./resources/dataset_train.csv')
data = data_raw.to_dict()
```

```python
def column_count(feature):
    column = data[feature]
    count = 0
    for key,value in column.items():
        if str(value) != 'nan':
            count = count + 1
        else:
            continue
    return count
```

```python
def column_mean(feature):
    column = data[feature]
    count = 0
    total = 0
    for key,value in column.items():
        if str(value) != 'nan':
            count = count + 1
            total = total + value
        else:
            continue
    return total/count
```

```python
def standard_deviation(feature):
    column = data[feature]
    count = 0
    total = 0
    for key,value in column.items():
        if str(value) != 'nan':
            count = count + 1
            total = total + value
        else:
            continue
    mean = total/count

    variance = 0
    for key, value in column.items():
        if str(value) != 'nan':
            res = value - mean
            square = res*res
            variance = variance + square
        else:
            continue
    return (variance/(count-1))**(1/2)
```

```python
def column_minimum(feature):
    column = data[feature]
    mini = np.inf
    for key,value in column.items():
        if mini > value:
            mini = value
        else:
            continue
    return mini

def column_maximum(feature):
    column = data[feature]
    maxi = - np.inf
    for key,value in column.items():
        if maxi < value:
            maxi = value
        else:
            continue
    return maxi
```

```python
def quartiles(feature, quart):
    mylist = list(data[feature].values())
    func_list = mylist.copy()
    func_list = [x for x in func_list if str(x) != 'nan']
    func_list.sort()
    quart_idx = quart*len(func_list)
    if quart_idx.is_integer() :
        result = (func_list[int(quart_idx)] + func_list[int(quart_idx)-1])/2
        return result
    else:
        result = func_list[int(np.floor(quart_idx))]
        return result
```

# Integration

```python
def create_describe(data) : # data = data_raw.to_dict()
    keys_float = [keys for keys in data.keys() if type(list(data[keys].values())[0]) == float]
    table = pd.DataFrame(columns=keys_float)
    table.loc['count'] = [column_count(keys) for keys in keys_float]
    table.loc['mean'] = [column_mean(keys) for keys in keys_float]
    table.loc['std'] = [standard_deviation(keys) for keys in keys_float]
    table.loc['min'] = [column_minimum(keys) for keys in keys_float]
    table.loc['25%'] = [quartiles(keys, 0.25) for keys in keys_float]
    table.loc['50%'] = [quartiles(keys, 0.5) for keys in keys_float]
    table.loc['75%'] = [quartiles(keys, 0.75) for keys in keys_float]
    table.loc['max'] = [column_maximum(keys) for keys in keys_float]
    
    return table
```

```python
create_describe(data)
```
